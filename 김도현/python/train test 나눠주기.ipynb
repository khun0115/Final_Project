## 셔플해서 0.2 만큼 이동

```python
MOVE_RATIO = 0.2

for label in label_list:
    folder_path = os.path.join(TRAIN_DATA_PATH, label)
    new_folder_path = os.path.join(TEST_DATA_PATH, label)
    
    file_list = os.listdir(folder_path)
    random.shuffle(file_list)
    
    move_num = int(len(file_list)*.2)
    print(label, move_num)
    print('========'*10)
    
    move_list = file_list[:move_num]
    
    for file in move_list:
        shutil.move(os.path.join(folder_path, file), os.path.join(new_folder_path, file))
    print(f'new {label} : ', len(os.listdir(new_folder_path)))
```
