```python
import os
import numpy as np
import shutil
```

# 본넷

```python
root = "C:\\Users\\acorn\\Desktop\\car_part"

root_dir = os.path.join(root,"train")
root_train = os.path.join(root_dir,"Bonnet")
test_root = os.path.join(root,"test\\Bonnet")
scratch_root = os.listdir(root_train)

for i in range(33):
    
    src = os.path.join(root_train,scratch_root[i])
    dst = os.path.join(test_root,scratch_root[i])
    
    shutil.move(src,dst)
    print("파일{}이 이동되었습니다".format(scratch_root[i]))
```
